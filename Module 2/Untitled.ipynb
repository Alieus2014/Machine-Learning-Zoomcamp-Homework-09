```python
df= pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv')
```

```python
df[["engine_displacement","horsepower","vehicle_weight","model_year","fuel_efficiency_mpg"]].head()
```

```python
s = df['fuel_efficiency_mpg'].dropna()
sk = s.skew()
print(f"Skew = {sk:.3f}")
print("Conclusion:",
      "long right tail" if sk > 1
      else "long left tail" if sk < -1
      else "no pronounced long tail")
```
